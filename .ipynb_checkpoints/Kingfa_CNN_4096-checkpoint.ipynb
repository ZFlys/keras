```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kingfa_datasets_path = "./kingfadatasets/"
```

```python
X = np.array(pd.read_csv(kingfa_datasets_path+"data.csv", index_col=0))
```

```python
y = np.array(pd.read_csv(kingfa_datasets_path+"target.csv", index_col=0))
```

```python
print(X.shape)
print(y.shape)
```

```python
import keras
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1024)
```

```python
y_train = keras.utils.to_categorical(y_train, 3)
y_test = keras.utils.to_categorical(y_test, 3)
```

```python
model = Sequential()
```

```python
model.add(Dense(1024, activation='sigmoid', input_shape=(4096,)))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(128, activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(3, activation='sigmoid'))
```

```python
adam = Adam(lr=1e-4)
```

```python
model.summary()
```

```python
model.compile(optimizer=adam, loss='squared_hinge', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=300, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test)
print("LOSS:", score[0])
print("Accuracy:", score[1])
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
